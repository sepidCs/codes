```python
import pandas as pd
```

```python
df = pd.read_excel("RMSE-istanbul+stock+exchange_0.9_iters_staticpr=10000.xlsx", index_col=0)
display(df)
df = df[:-4]
df.rename(columns={"no feature selection": "None"}, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Create a box plot
df.boxplot(figsize=(10, 6))

# Add title and labels
plt.title("Box Plot of RMSE Across Methods for energy = 0.90")
plt.ylabel("RMSE")
plt.xticks(rotation=45)
plt.savefig("boxplot_Istanbul_90.png", dpi=300, bbox_inches='tight')

plt.show()
```
