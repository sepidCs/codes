```python
import pandas as pd
```

```python
df = pd.read_excel("RMSE-appliances+energy+prediction_0.95_iters_staticpr=10000.xlsx", index_col=0)

df = df[:-4].drop(columns=["TSFRESH"])
df.rename(columns={"no feature selection": "None"}, inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Create a box plot
df.boxplot(figsize=(10, 6))

# Add title and labels
plt.title("Box Plot of RMSE Across Methods for energy = 0.95")
plt.ylabel("RMSE")
plt.xticks(rotation=45)
plt.savefig("boxplot_energy_95.png", dpi=300, bbox_inches='tight')

plt.show()
```
